### Indexing and Slicing

```python
df = pd.read_csv('/Users/lorisegovia/Downloads/Employees.csv')
```

```python
#return index label 5 using loc
df.loc[5]
```

```python
#return multiple indexes
#another set of brackets to hold a list
df.loc[[2,4,7]]
```

```python
#range of indexes
df.loc[3:5]
```

```python
#only specified range of columns for a specified range of indexes
df.loc[0:2, "Phone":"Skills"]
```

```python
#address to be table index
df.set_index("Address", inplace=True)
df
```

```python
#full rows returned for specified range of values
df.loc["1st Address, Miami":"5th Address, Miami"]
```

```python
#only want phone
df.loc["1st Address, Miami":"5th Address, Miami", "Phone"]
```

```python
#multiple columns for specific range
df.loc["1st Address, Miami":"5th Address, Miami", ["LastName", "Skills"]]
```

```python
#range of columns associated with a specified range of values
df.loc["1st Address, Miami":"5th Address, Miami", "LastName":"Skills"]
```

```python
#grab single number
df.loc["1st Address, Miami", "Phone"]
```

```python
#phone value for all entries : means all
df.loc[:, "Phone"]
```

```python
#all columns for all rows means whole table
df.loc[:,:]
```

```python
#return unique values
set(df.loc[:, "Department"])
```

```python
#position based locating is iloc
#slicing is done the old python way
#second index will not be included in the slice
#indexing starts at zero
#we get 4
df.iloc[4]
```

```python
#want 5,7,9
df.iloc[[4,6,8]]
```

```python
#columns at 2,3,4 means that 5 is upper limit and isnt included in slice
# 2 is the column we want to start at
df.iloc[2, 2:5]
```

```python
#print out 2,3,4
df.iloc[:,2:5]
```

```python

df = pd.read_csv('/Users/lorisegovia/Downloads/Employees.csv')
```

```python
df.set_index('ID', inplace=True)
```

```python
#salary less than 50k
df[df.loc[:,"Salary"] < 50000]
```

```python
#less than 50 or greater than 56
df[(df.loc[:,"Salary"] < 50000) | (df.loc[:,"Salary"] > 56000)]
```

```python
# > 50 and from sales department
df[(df.loc[:,"Salary"] > 50000) & (df.loc[:,"Department"] == "Sales")]
```

```python
# allemployes who are not in it department
df[~(df.loc[:, "Department"] == "IT")]
```
