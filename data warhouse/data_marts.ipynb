- same person, diff spellings

## Data Warehouse Features
- provides strategic enterprise level decision support
- has multi-dimensional view on the enterprise data
- caters to the entire spectrum of management
- uses descriptive and standard business terms
- has high degree of scalability
- possesses high analytical capability
- stores historical data

    - campaigns
- profit trends
- procurement analysis
- statistical analysis
- claims analysis
- manufacturing optimization
- customer relationship management

## Components of a Data Warehouse
- staging area
    - a preparatory repository where transaction data can be transformed for use in the data warehouse
    - created to aid in the collection and transformation of datqa that is targeted for a data warehouse
- data warehouse
    - data integrated from various sources are stored here
    - dimensional modeling is extensively used in data warehousing
    - enterprised data warehouses will typically follow normalized data models
    - data warehouse is used on an enterprise level
    - data warehouses are arranged around the corporate subject areas found in the corporate data model
- data mart
    - traditional dimensionally modeled set of dimension and fact tables
    - per kimball, a data warehouse is the union of a set of data marts
        - data marts are typically used on a business division / department level
- business intelligence tools
    - a set of tools that allow uses to access enterprise data via reports, online analytical processing (OLAP) cubes, graphs/charts, as-hoc queries and dashboards

## Data Warehouse Modeling
- data warehouses typically use a de-normalized methof call dimensional modeling
    - dimension:  
        - an entity defined in its entirety with a single primary key
        - examples:  customer, product, sales force, calendar
    - fact:
        - details (often numerical) regarding a set of dimensions
        - example:  order details
