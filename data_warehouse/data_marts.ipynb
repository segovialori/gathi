## Components of a Data Warehouse
- staging area
    - a preparatory repository where transaction data can be transformed for use in the data warehouse
    - created to aid in the collection and transformation of datqa that is targeted for a data warehouse
- data warehouse
    - data integrated from various sources are stored here
    - dimensional modeling is extensively used in data warehousing
    - enterprised data warehouses will typically follow normalized data models
    - data warehouse is used on an enterprise level
    - data warehouses are arranged around the corporate subject areas found in the corporate data model
- data mart
    - traditional dimensionally modeled set of dimension and fact tables
    - per kimball, a data warehouse is the union of a set of data marts
        - data marts are typically used on a business division / department level
- business intelligence tools
    - a set of tools that allow uses to access enterprise data via reports, online analytical processing (OLAP) cubes, graphs/charts, as-hoc queries and dashboards

## Data Warehouse Modeling
- data warehouses typically use a de-normalized method call dimensional modeling
    - dimension:  
        - an entity defined in its entirety with a single primary key
        - primary key defines whats inside
        - examples:  customer, product, sales force, calendar
    - fact:
        - what you really what to measure
        - facts are surrounded by dimensionr
        - details (often numerical) regarding a set of dimensions
        - example:  order details
        - one record would be a sale for a given customer on a given day
        


## BI Components
- Reports
    - Commonly needed data can be structured in a set of canned reports made available to large numbers of users
    - Flexibility can be given to users through ad-hoc querying and filters


- Cubes
    - Multi-dimensional, allowing the user to view the data from multiple angles
- Interactive, giving the user the ability to change what is viewable on the fly

- Charts and Graphs
    - Graphical Representation of data
    - Commonly used in presentations and statistical analysis

- Dashboards
    - Actively updating graphical displays that provides business users with updates on key metrics
    - Some dashboard provide drill through capability, allowing users to start with summary data and dive into the details


## What is MDM?
- The collection of practices and technologies to
    - Sanitize
    - Standardize
    - Integrate
    - Create and maintain
- Core enterprise data for
    - Customers
    - Suppliers
    - Products
    - Chart of accounts
    - Other business entities
- Existing in disparate source systems such as ERP, CRM, etc.
- To provide a Single True View


## What does MDM involve?
- MDM integration involves 
    -Data collection
    - Data modeling
    - Data quality
    - Data Standardization
    - Data Integration
    - A Single True View
- Provides consistent information framework that enables enterprises to 
    - Remain responsive
    - Accurately measure performance
    - Operate Efficiently


## Benefits of MDM
- Consolidates enterprise wide data without the time, cost and risk involved in integrating multiple applications
- Provides a Single True View of the master data
- Decision-making based on accurate, complete, uniform and timely data allowing enhanced
    - Financial analysis and planning capabilities
    - Accurate performance measurement and reporting
    - Cross-system analysis
    - Ability to respond quicker to dynamic markets
    - Operational efficiency
    - Marketing campaigns
    - Integration from M&A
- Effective data management (separated from the business processes and functional systems) for the long term with reduced code of data maintenance
- Helps to improve proximity with customers, vendors, suppliers, regulators, and other stakeholders
- A step towards attaining competitive advantage


## Types of Metadata

- Business: Descriptive information about the reports, transformation, tables, columns, etc. for the business users
- Technical: Physical characteristics of the data such as table/column names, data types, sources to targets, mappings, etc. for the technical team
- Operational: Allows admins to monitor operations and usage of the data warehouse (load times, load statistics, etc.)

## Advantages of Metadata

- The advantages of building a metadata repository are
    - Ease of maintenance of the data warehousing environment
    - Ease in enforcement of standards
    - More central managed data, avoiding inconsistency problems
    - Greater integration of information from multiple subject areas
    - Impact analysis capability
        - Impact analysis of the changes in the source on DWH, hence the OLAP / data mining reports
    - Ultimately, enables data to become information
- When you know what data you have, you can trust it
